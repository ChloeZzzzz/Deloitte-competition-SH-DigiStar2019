```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import seaborn as sns
import matplotlib.font_manager as fm
```

```python
data = pd.read_csv("财务维度打分.csv")
data = data.rename(columns=data.iloc[1, :])
data = data.iloc[2:, (data.shape[1] - 3):]
data
```

```python
labels = np.array(['研发维度', '生产成本维度', '财务风险维度'])
#for i in range(data.shape[0]):
datadf = data.iloc[i].values

for i in range(len(datadf)):
    datadf[i] = float(datadf[i])

angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
datadf = np.concatenate((datadf,[datadf[0]]))
angles = np.concatenate((angles,[angles[0]]))
font = fm.FontProperties(fname='/System/Library/Fonts/STHeiti Light.ttc')

fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, datadf, 'o-', linewidth=2)
ax.fill(angles, datadf, alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties=font)
ax.grid(True)
```
