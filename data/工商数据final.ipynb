```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import matplotlib.font_manager as fm
```

```python
data = pd.read_excel("工商数据final.xlsx")
data = data.rename(columns={"A": "注册资本（亿）", "B": "品牌数量", "C": "成立时间（年）", "D": "雇员人数"})
data
```

```python
#for i in range(1, data.shape[0]):
#输入公司前的序号，eg：三全食品，3；黑芝麻，1
newdata = pd.DataFrame(data.iloc[3, 2:6])

font = fm.FontProperties(fname='/System/Library/Fonts/STHeiti Light.ttc')  # speicify font
ax = newdata.plot(kind='bar')
ax.set_xticklabels(labels=newdata.index, fontproperties=font)
ax.legend("")
ax.set_title("企业基本信息画像", fontproperties=font)
plt.show()
```
