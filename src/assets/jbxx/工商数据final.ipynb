```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import matplotlib.font_manager as fm
%matplotlib inline
plt.rcParams['font.family'] = ['Arial Unicode MS']
#plt.rcParams['font.sans-serif']=['SimHei']
#plt.rcParams['axes.unicode_minus']=False
```

```python
data2 = pd.DataFrame(pd.read_excel("食品上市公司名单2.xlsx"))
data2 = data2.rename(columns=data2.iloc[0,:])
data2 = data2.iloc[1:, 9:]
data2
```

```python
data = pd.read_excel("工商数据final.xlsx")
data = data.rename(columns={"Table Data 4": "市公司简称", "A": "注册资本（亿）", "B": "品牌数量", "C": "成立时间（年）", "D": "雇员人数"})
data = data.iloc[1:,:]
data = pd.merge(data, data2, on="市公司简称")
data
```

```python
#for i in range(1, data.shape[0]):
#输入公司前的序号，eg：三全食品，3；黑芝麻，1
for i in range(0,43):
    newdata = pd.DataFrame(data.iloc[i, 2:6])

    #font = fm.FontProperties(fname='/System/Library/Fonts/STHeiti Light.ttc')  # speicify font
    ax = newdata.plot(kind='bar')
    plt.rcParams['figure.figsize'] = (5.0, 4.0)
    ax.set_xticklabels(labels=newdata.index)
    ax.set_xticklabels(list(newdata.index), rotation=0)
    #, fontproperties=font , fontproperties=font
    ax.legend("")
    plt.rcParams['savefig.dpi'] = 200 
    ax.set_title("基本信息画像")
    fig = ax.get_figure()
    fig.savefig(r'/Users/wangning/Desktop/项目素材/jbxx/%s.jpg'%data.iloc[i,25])
```
