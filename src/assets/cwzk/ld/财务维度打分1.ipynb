```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import seaborn as sns
import matplotlib.font_manager as fm
%matplotlib inline
plt.rcParams['font.family'] = ['Arial Unicode MS']
```

```python
data = pd.read_csv("财务维度打分.csv")
data = data.rename(columns=data.iloc[1, :])
data1= data.iloc[2:,1]
data = data.iloc[2:, (data.shape[1] - 3):]
data1 = pd.DataFrame(data1)
print(data1)
data2 = pd.DataFrame(pd.read_excel("食品上市公司名单2.xlsx"))
data2 = data2.rename(columns=data2.iloc[0,:])
data2 = data2.iloc[1:, 9:]

data1.columns = ["市公司简称"]
data1 = pd.merge(data1, data2, on="市公司简称")
data1
```

```python
#for i in range():

for i in range(0,data.shape[0]-1):
    labels = np.array(['研发维度', '生产成本维度', '财务风险维度'])
    datadf = data.iloc[i].values
    k=i

    for i in range(len(datadf)):
        datadf[i] = float(datadf[i])

    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
    datadf = np.concatenate((datadf,[datadf[0]]))
    angles = np.concatenate((angles,[angles[0]]))

    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True)
    ax.plot(angles, datadf, 'o-', color='yellow')
    ax.fill(angles, datadf, alpha=0.25)
    ax.set_thetagrids(angles * 180/np.pi, labels)
    ax.grid(True)
    plt.rcParams['savefig.dpi'] = 200
    fig=ax.get_figure()
    fig.savefig(r'/Users/wangning/Desktop/项目素材/cwzk/ld/%s.jpg'%data1.iloc[k,1])
```
