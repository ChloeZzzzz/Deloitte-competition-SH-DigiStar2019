```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import matplotlib.font_manager as fm
%matplotlib inline
plt.rcParams['font.family'] = ['Arial Unicode MS']
#plt.rcParams['font.sans-serif']=['SimHei']
#plt.rcParams['axes.unicode_minus']=False
```

```python
data = pd.read_excel("z-score label2.xlsx")
data = data.rename(columns={"Unnamed: 0": "市公司简称"})
data2 = pd.DataFrame(pd.read_excel("食品上市公司名单2.xlsx"))
data2 = data2.rename(columns=data2.iloc[0,:])
data2 = data2.iloc[1:, 9:]

datadf = []
for i in range(0, data.shape[0]):
    datadf.append(data.iloc[i,:])
    
datadf = pd.DataFrame(datadf)
datadf = pd.merge(datadf, data2, on="市公司简称")
datadf
```

```python
for i in range(0,42):
    newdata = pd.DataFrame(datadf.iloc[i, 1:8])
    index = ['']
    for j in newdata.index:
        index.append(j)

    #font = fm.FontProperties(fname='/System/Library/Fonts/STHeiti Light.ttc')  # speicify font
    ax = newdata.plot(kind='line', style='.-')
    ax.set_xticklabels(labels=index)
    ax.legend("")
    #ax.set_ylim(bottom=0, top=1500)
    plt.rcParams['savefig.dpi'] = 200
    plt.rcParams['figure.figsize'] = (6.5, 6.5)
    ax.set_title("近三年财务风险评估")
    plt.axhline(y=1.23, color='gray', linestyle='--')
    plt.axhline(y=2.9, color='gray', linestyle='--')
    fig=ax.get_figure()
    fig.savefig(r'/Users/wangning/Desktop/项目素材/cwzk/Zscore/%s.jpg'%datadf.iloc[i,8])
```
