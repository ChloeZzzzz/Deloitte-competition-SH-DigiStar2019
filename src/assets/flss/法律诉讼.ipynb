```python
data = pd.read_excel("法律诉讼final.xlsx")
data = data.rename(columns=data.loc[0])
data = data.iloc[1:,:]
data2 = pd.DataFrame(pd.read_excel("食品上市公司名单2.xlsx"))
data2 = data2.rename(columns=data2.iloc[0,:])
data2 = data2.iloc[1:, 9:]
data = pd.merge(data, data2, on="市公司简称")
data
```

```python
#for i in range(1, data.shape[0]):
for i in range(1,44):
    newdata = pd.DataFrame(data.iloc[i, 3:15])
    index = [""]
    for j in newdata.index:
        index.append(j)

    #font = fm.FontProperties(fname='/System/Library/Fonts/STHeiti Light.ttc')  # speicify font
    ax = newdata.plot(kind='line', style='.-')
    ax.set_xticklabels(labels=index)
    ax.legend("")
    #ax.set_ylim(bottom=0, top=1500)
    ax.set_title("法律诉讼趋势图")
    plt.rcParams['savefig.dpi'] = 200
    fig=ax.get_figure()
    fig.savefig(r'/Users/wangning/Desktop/项目素材 /法律诉讼/%s.jpg'%data.iloc[i,23])
```
